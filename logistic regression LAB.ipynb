```python
+
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('IRIS-Copy1.csv')
df.head(10)
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df.species.unique()
```

```python
df=df[df['species']!='Iris-setosa']
df.head(20)
```

```python
df['species'].value_counts()
```

```python
df['species']=df['species'].map({'Iris-versicolor':1,'Iris-virginica':0})
```

```python
df.head(60)
```

```python
x=df.drop('species',axis=1)
x.head()
```

```python
y=df['species']
y.head()
```

## creating the model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lo=LogisticRegression()
```

## doing the hyperparameter tuning very very important

```python
from sklearn.model_selection import GridSearchCV
parameter={'penalty':['l1','l2','elasticnet'],'C':[2,5,6,7,50,62,78],'max_iter':[100,200,300]}
```

```python
classifier=GridSearchCV(lo,param_grid=parameter,scoring='accuracy',cv=5)
classifier
```

```python
classifier.fit(x_train,y_train)
```

## to check out of the given parameter values which one  is selected

```python
print(classifier.best_params_)
```

```python
print(classifier.best_score_)
```

### now performing predictions

```python
y_pred=classifier.predict(x_test)
y_pred
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(y_pred,y_test))
```

```python
metrics.accuracy_score(y_test,y_pred)
```
