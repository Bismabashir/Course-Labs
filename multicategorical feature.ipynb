```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('mercedesbenz.csv',usecols=['X1','X2','X3','X4','X5','X6'])
df.head(10)
```

### checking the unique categories present in each column

```python
for col in df:
    print(col,':',len(df[col].unique()),'labels')
```

```python
pd.get_dummies(df,drop_first=True)
```

## checking the top 10 categories in X1 column

```python
df.X1.value_counts().head(10)
```

## making a list of these top 10 categories

```python
top10=[i for i in df.X1.value_counts().head(10).index]
top10
```

## applying the technique of converting these top categories into binary

```python
for label in top10:
    df[label]=np.where(df['X1']==label,1,0)
df[['X1']+top10].head(20)
```

## making a function to make it easy for converting it into binary for all the features

```python
def on_hot(dataframe,columns,top10):
    for label in top10:
        df[columns+'_'+label]=np.where(df[columns]==label,1,0)
on_hot(df,'X1',top10)
df.head()
```

```python
top10=[i for i in df['X2'].value_counts().head(10).index]
print(top10)
on_hot(df,'X2',top10)
df.head()
```

```python
top10=[i for i in df['X3'].value_counts().head(10).index]
print(top10)
on_hot(df,'X3',top10)
df.head()
```
